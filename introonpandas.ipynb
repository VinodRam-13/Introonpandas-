```python
# prompt: can you calculate measure of central tendency for salary columns and measures of dispersion for salary columns, also calculate skewness and kutosis

import pandas as pd

# Load the dataset (assuming it's in the current working directory or provide the correct path)
df = pd.read_csv('/content/Salaries.csv')

# Calculate measures of central tendency for the 'salary' column
salary_mean = df['salary'].mean()
salary_median = df['salary'].median()
salary_mode = df['salary'].mode()

# Calculate measures of dispersion for the 'salary' column
salary_variance = df['salary'].var()
salary_std_dev = df['salary'].std()
salary_range = df['salary'].max() - df['salary'].min()  # Range

# Calculate skewness and kurtosis for the 'salary' column
salary_skewness = df['salary'].skew()
salary_kurtosis = df['salary'].kurt()

# Print the results
print("Measures of Central Tendency:")
print(f"Mean Salary: {salary_mean}")
print(f"Median Salary: {salary_median}")
print(f"Mode Salary: {salary_mode.iloc[0]}") # Print first mode if there are multiple

print("\nMeasures of Dispersion:")
print(f"Variance: {salary_variance}")
print(f"Standard Deviation: {salary_std_dev}")
print(f"Range: {salary_range}")

print("\nSkewness and Kurtosis:")
print(f"Skewness: {salary_skewness}")
print(f"Kurtosis: {salary_kurtosis}")
```
